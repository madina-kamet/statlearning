```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_digits, load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from numpy import mean
from numpy import std
from sklearn.datasets import make_classification, make_blobs
from sklearn.model_selection import cross_val_score
from matplotlib import pyplot
from sklearn.metrics import accuracy_score
plt.rcParams['figure.figsize'] = [15, 5]
```

```python
data = load_breast_cancer()

# Store the feature data
X = data.data
# # store the target data
y = data.target
# split the data using Scikit-Learn's train_test_split
from sklearn.model_selection import train_test_split


test_score_array = []
train_score_array = []



for d in range(26):
  
  x_local = X[:, d:]

  x_train, x_test, y_train, y_test = train_test_split(x_local, y, train_size=0.7)

  model = LogisticRegression(max_iter=10000000000)
  model.fit(x_train, y_train)

  y_pred_test = model.predict(x_test)


  train_score_array.append(1-model.score(x_train, y_train))

  test_score_array.append(1-accuracy_score (y_test, y_pred_test))
  if(d in [0, 10, 20, 25]):
    print('___________________________________________________________')
    print()
    print('Train error for %s: %.3f '% (30-d, 1-model.score(x_train, y_train)))
    print('Test error for %s: %.3f ' % (30-d, 1-accuracy_score (y_test, y_pred_test)))
    print()


plt.subplot(1, 2, 1)
plt.plot(test_score_array)
plt.title("Test error for changing d")




plt.subplot(1, 2, 2)
plt.plot(train_score_array)
plt.title("Train error for changing d")


plt.show()
```
