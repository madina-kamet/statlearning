```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_digits, load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from numpy import mean
from numpy import std
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.pipeline import Pipeline
from numpy import mean
from numpy import std
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from matplotlib import pyplot
from sklearn.metrics import accuracy_score
plt.rcParams['figure.figsize'] = [15, 5]
```

```python
data = load_breast_cancer()

# # Import pandas
# import pandas as pd
# # Read the DataFrame, first using the feature data
# df = pd.DataFrame(data.data, columns=data.feature_names)
# # Add a target column, and fill it with the target data
# df['target'] = data.target
# # Show the first five rows
# df.head()


# Store the feature data
X = data.data
# store the target data
y = data.target
# split the data using Scikit-Learn's train_test_split
from sklearn.model_selection import train_test_split




test_score_array = []
train_score_array = []



for d in range(26):
  
  x_local = X[:, d:]

  x_train, x_test, y_train, y_test = train_test_split(x_local, y, train_size=0.7)

  model = LogisticRegression(max_iter=10000000000)
  model.fit(x_train, y_train)

  y_pred_test = model.predict(x_test)


  train_score_array.append(model.score(x_train, y_train))

  test_score_array.append(accuracy_score (y_test, y_pred_test))
  if(d in [0, 10, 20, 25]):
    print('___________________________________________________________')
    print()
    print('Train for %s: %.3f '% (30-d, model.score(x_train, y_train)))
    print('Test for %s: %.3f ' % (30-d, accuracy_score (y_test, y_pred_test)))
    print()


plt.subplot(1, 2, 1)
plt.plot(test_score_array)
plt.title("Test score Array")




plt.subplot(1, 2, 2)
plt.plot(train_score_array)
plt.title("Train score Array")


plt.show()
```

```python
data = load_breast_cancer()

# # Import pandas
# import pandas as pd
# # Read the DataFrame, first using the feature data
# df = pd.DataFrame(data.data, columns=data.feature_names)
# # Add a target column, and fill it with the target data
# df['target'] = data.target
# # Show the first five rows
# df.head()


# Store the feature data
X = data.data
# store the target data
y = data.target
# split the data using Scikit-Learn's train_test_split
from sklearn.model_selection import train_test_split




test_score_array = []
train_score_array = []



for m in range(1,10):
  
  x_local = X[:, :]

  x_train, x_test, y_train, y_test = train_test_split(x_local, y, train_size=0.1*m)

  model = LogisticRegression(max_iter=10000000000)
  model.fit(x_train, y_train)

  y_pred_test = model.predict(x_test)


  train_score_array.append(model.score(x_train, y_train))

  test_score_array.append(accuracy_score (y_test, y_pred_test))
  print('___________________________________________________________')
  print()
  print('Train for %.1f: %.3f '% (0.1*m, model.score(x_train, y_train)))
  print('Test for %.1f: %.3f ' % (1-0.1*m, accuracy_score (y_test, y_pred_test)))
  print()


plt.subplot(1, 2, 1)
plt.plot(test_score_array)
plt.title("Test score Array")




plt.subplot(1, 2, 2)
plt.plot(train_score_array)
plt.title("Train score Array")


plt.show()
```
