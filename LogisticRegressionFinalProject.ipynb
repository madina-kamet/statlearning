```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_digits
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

plt.rcParams['figure.figsize'] = [15, 5]
```

```python
test_score_array = []
train_score_array = []
for d in range(21):
  x, y = load_digits(return_X_y=True)
  x = x[:,d:]
  x_train, x_test, y_train, y_test =\
      train_test_split(x, y, test_size=0.4, random_state=0)

  # используем чтобы данные подходили для logistic regression
  scaler = StandardScaler()
  x_train = scaler.fit_transform(x_train)


  model = LogisticRegression(solver='liblinear', C=1, multi_class='ovr',
                            random_state=0)
  model.fit(x_train, y_train)

  y_pred_test = model.predict(x_test)


  train_score_array.append(1-model.score(x_train, y_train))

  test_score_array.append(1-model.score(x_test, y_test))



plt.subplot(1, 2, 1)
plt.plot(test_score_array)
plt.title("Test score Array")




plt.subplot(1, 2, 2)
plt.plot(train_score_array)
plt.title("Train score Array")


plt.show()
```
